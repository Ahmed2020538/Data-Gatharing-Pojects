```python
# Import requier Liberaries ::
#-------------------------------
from selenium import webdriver
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager as CDM
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from openpyxl import load_workbook
from time import sleep
import csv
import sqlite3 as S
import pandas as pd
import numpy as np
```

```python
# Scraping Script ::

number = 1
url = 'http://www.arab-pa.org/en/shared/Publishers.aspx'

# Install Chrome Maneger  and get Response ::
driver = webdriver.Chrome(CDM().install())
driver.get(url)

# Create CSV to inset data on it ::
with open("Arab-Paa.csv" , "w+") as f :
    
    writer = csv.DictWriter(f , fieldnames=["Publisher" ,
                                            "Name" , "Job" , "Country" ,
                                            "Phone" , "Fax" , "Mobile" , "E_mail" , "WebSite"])
    writer.writeheader()
    # Create DataBase to insert data on it also ::
    con =  S.connect("Arab-Paa.db")
    con.execute("""create table AraBiaa_details ( Publisher str, Name str, Job str, Country str ,
    
                    Phone str , Fax str, Mobile str , E_mail str , WebSite str)""")
    
    # Scraping data
    while number <= 135:
        
        # get Div include all deatails we need ::
        containers = driver.find_elements_by_css_selector(".pricing-desc")
        
        for container in containers:
            
        # puplisher = [container.find_element_by_css_selector("div.pricing-title h3").text if
        # len(container.find_elements_by_css_selector("div.pricing-title h3")).text) > 0 else np.nan ] 
            # inested of try and expect statement
            
            try:
                publisher = container.find_element_by_css_selector("div.pricing-title h3").text
            except:
                punlisher = np.nan
            try:
                name = container.find_element_by_css_selector(".iconlist-color li:nth-child(1)").text.split(":")[1]
            except:
                name =  np.nan
            try:
                job = container.find_element_by_css_selector(".iconlist-color li:nth-child(2)").text.split(":")[1]
            except:
                job =  np.nan
            try:
                country = container.find_element_by_css_selector(".iconlist-color li:nth-child(3)").text.split(":")[1]
            except:
                country =  np.nan
            try:
                phone = container.find_element_by_css_selector(".iconlist-color li:nth-child(4)").text.split(":")[1]
            except:
                phone =  np.nan
            try:
                fax = container.find_element_by_css_selector(".iconlist-color li:nth-child(5)").text.split(":")[1]
            except:
                fax =  np.nan
            try:
                mobile = container.find_element_by_css_selector(".iconlist-color li:nth-child(6)").text.split(":")[1]
            except:
                mobile =  np.nan
            try:
                email = container.find_element_by_css_selector(".iconlist-color li:nth-child(7)").text.split(":")[1]
            except:
                email =  np.nan
            try:
                website = container.find_element_by_css_selector(".iconlist-color li:nth-child(8)").text.split(":")[1]
            except:
                website =  np.nan
                
                
            # insert data in csv file and Database ::
            writer.writerow({"Publisher" : publisher , "Name" : name , "Job" : job ,
                             "Country" : country , "Phone" : phone ,
                             "Fax" : fax , "Mobile" : mobile , "E_mail" : email , "WebSite" : website } )
            sql  = """insert into AraBiaa_details (Publisher , Name , Job , Country ,
                      Phone , Fax , Mobile , E_mail , WebSite) values(?,?,?,?,?,?,?,?,?)"""
            data = [ publisher , name , job , country , phone , fax , mobile , email , website]
            con.execute(sql , data)
            con.commit()
    WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//table[@id="ctl00_ContentPlaceHolder1_GridView1"]//table/tbody/tr//td/span//following::td/a'))).click()
    print(number)
    number += 1
driver.quit()
```
