# 1 - Func To get all Books Data :

```python
def get_books_data(url) :
    with open("Books_All_Details.csv" , "w+") as f :
        writer = csv.DictWriter( f , fieldnames=["Title" , "Rateing" , "Price" , "status" ])
        writer.writeheader()
        res = R.get(url)
        soup = B(res.text , "html.parser")
        data = soup.find_all("article"  , attrs={"class" : "product_pod"})
        Rating = {"One" : 1 , "Two" : 2 , "Three" : 3 , "Four" : 4 , "Five" : 5}
        for d in data :
            Rate = d.find("p").get("class")[1]
            Title = d.find("h3").get_text()
            Price = d.find("p" , attrs = {"class" : "price_color"}).get_text().split("Â£")[1]
            status = d.find( "p", attrs = {"class" : "instock availability"}).get_text().strip()
            writer.writerow({"Title" : Title, "Rateing" : Rate, "Price" : Price , "status" : status})
            print(f"Rate :: {Rating[Rate]}\nTitle :: {Title}\nPrice :: {Price}\nstatus{status}\n======")
```

# 3 - Func To get Books Names only

```python
def get_books_Name(url) :
    Books_name = []
    with open("Books_All_Details.csv" , "w+") as f :
        writer = csv.DictWriter( f , fieldnames=["Title" , "Rateing" , "Price" , "status" ])
        writer.writeheader()
        res = R.get(url)
        soup = B(res.text , "html.parser")
        data = soup.find_all("article"  , attrs={"class" : "product_pod"})
        for d in data :
            Title = d.find("h3").get_text()
            Books_name.append(Title)
            #print(f"Title :: {Title}\n")
    return Books_name
```

```python
get_books_Name("http://books.toscrape.com/catalogue/category/books/mystery_3/index.html")
```

# 4 - func to get Categories Name and Categories Links

```python
def get_categories(url) :
    with open("Categories.csv" , "w+" ) as f :
        categories = {}
        writer = csv.DictWriter( f , fieldnames=["Category" , "Category_Link"])
        writer.writeheader()
        res = R.get(url)
        soup = B(res.text , "html.parser")
        data = soup.find("div" , attrs={"class": "side_categories"}).find("ul").find("li").find("ul").find_all("a")
        for d in data :
            Category = d.get_text().strip()
            Link = d.get("href").replace("../" , "http://books.toscrape.com/catalogue/category/books/")
            if Link != "index.html":
                categories[Category] = Link
                
                writer.writerow({"Category" : Category , "Category_Link" :Link})
                #print(f"Category :: {Category}\nLink :: {Link}")
        return categories
```

```python
get_categories("http://books.toscrape.com/catalogue/category/books/mystery_3/index.html")
```
