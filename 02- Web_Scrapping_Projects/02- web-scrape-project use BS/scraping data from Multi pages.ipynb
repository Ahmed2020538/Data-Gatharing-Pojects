```python
import requests as R
import pandas as pd
import csv
from bs4 import BeautifulSoup as B
```

```python
url = "https://egypt.souq.com/eg-en/souqegypt/speakers-22/a-t/p/?_=1586878023591&ref=nav&section=2&page=1"
res = R.get(url)
soup = B(res.text , "html.parser")
data = soup.find_all("div" , attrs={"class" , "columns small-7 medium-12"})
for d in data :
    title = d.find("h6").get_text().strip()
    orignal_price = d.find("h5" , attrs={"class" : "price"}).get_text().split("EGP")[1].strip()
    Current_price = d.find("h5" , attrs={"class" : "price"}).get_text().split("EGP")[0].strip()
    Shipping = d.find("strong").get_text()
    
    print(f"The Describe of Product :: {title}")
    print(f"The orignal_price :: {orignal_price}")
    print(f"The Current_price :: {Current_price}")
    print(f"The Shipping :: {Shipping}")
    print("="*50)
    print("="*50)
```

```python
def get_data(url) :
    with open("data.csv" , "w+") as f:
        
        writer = csv.DictWriter(f , fieldnames=["title" , "orignal_price" , "Current_price" , "Shipping"])
        writer.writeheader()
        res = R.get(url)
        soup = B(res.text , "html.parser")
        data = soup.find_all("div" , attrs={"class" , "columns small-7 medium-12"})
        for d in data :
            title = d.find("h6").get_text().strip()
            orignal_price = d.find("h5" , attrs={"class" : "price"}).get_text().split("EGP")[1].strip()
            Current_price = d.find("h5" , attrs={"class" : "price"}).get_text().split("EGP")[0].strip()
            Shipping = d.find("strong").get_text()
            writer.writerow({"title" :title , "orignal_price" :orignal_price , "Current_price" : Current_price , "Shipping" : Shipping})
            print(f"The Describe of Product :: {title}")
            print(f"The orignal_price :: {orignal_price}")
            print(f"The Current_price :: {Current_price}")
            print(f"The Shipping :: {Shipping}")
            print("="*50)
            print("="*50)
```

```python
get_data("https://egypt.souq.com/eg-en/souqegypt/speakers-22/a-t/p/?_=1586878023591&ref=nav&section=2&page=1")
```
