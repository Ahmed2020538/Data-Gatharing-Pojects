```python
# Initialization driver-browser and set some preferances and options to driver
# To make it easy for crawling or scraping data
def init_driver(gecko_driver = '' , user_agent = '' , load_image = True , is_headless = False ) :
    
    # set common behaviour Preferances ::
    # define firfox_profit
    firfox_profile = WD.FirefoxProfile()
    # 1 - disable flashplayer
    firfox_profile.set_preference('dom.ipc.plugins.enabled.libflashplayer.so', False)
    # 2 - Mute Voice 
    firfox_profile.set_preference('media.volume_scale', '0.0')
    # 3 - ignore Web-Notification to avoid crash browser
    firfox_profile.set_preference("dom.webnotifications.enabled", False)
    # 4 - ignore load image to save bandwith and make time processing
    if not load_image : # false
        firfox_profile.set_preference('permissions.default.image', 2) # load = 1 and dis = 2
        
    # 5 - set use_agent constant for any device
    if user_agent != '' :
        firfox_profile.set_preference('general.useragent.override', user_agent)
        
        
    # Set firfox Options
    
    options = Options()
    
    
    # to make browser load with devices and servers with or without GUI
    # If True without GUI If False with
    options.headless = is_headless
    
    
    
    
    
    
    driver = WD.Firefox(executable_path = f"{current_path}/{gecko_driver}",
                        firefox_profile = firfox_profile ,
                        options = options )
    
    
    
    return driver
```

```python
# load Firefox_Browser automated By initialization function
driver = init_driver(gecko_driver = gecko_driver , user_agent = user_agent)
```

```python

cats = ["phones-tablets" ,"video-games", "computing" ,
        "electronics" , "home-office" , "category-fashion-by-jumia" , "sporting-goods"]
for cat in cats :
    cats_url = f"{org_jumia_url}/{cat}"
    for i in range(1,50) :
        page_url = f"{cats_url}?page={str(i)}#catalog-listing"
        
        # call func to get response of data    
        _ = get_url(page_url, driver )
        # Create csv to load data on it ::
        with open("Jumia_products_details.csv" , "w+" , encoding="utf-8") as f:
            writer = csv.DictWriter(f , fieldnames=["Name" , "Price_After_disc" , "Price_Before_disc" ,
                                                    "Product_disc" , "Product_evaluate" ,
                                                    "Product_link" , "Category"])
            writer.writeheader()
            try :
                products = driver.find_elements_by_class_name("prd")
                for product in products :
                    Name = [product.find_element_by_class_name("name").text if
                                         len(product.find_elements_by_class_name("name")) > 0 else np.nan ]
            
                    Price_After_disc  = [product.find_element_by_class_name("prc").text.split()[1]if
                                         len(product.find_elements_by_class_name("prc")) > 0 else np.nan ]
        
        
                    Price_Before_disc = [ product.find_elements_by_class_name("old")[0].text.split()[1] if
                                         len(product.find_elements_by_class_name("old")) > 0 else np.nan ]
    
                    Product_disc = [ product.find_elements_by_class_name("tag")[0].text.split('%')[0] if
                                         len(product.find_elements_by_class_name("tag")) > 0 else np.nan ]
    
                    Product_evaluate = [ product.find_elements_by_class_name("stars")[0].text.split()[0] if
                                         len(product.find_elements_by_class_name("stars")) > 0 else np.nan ]
    
                    Product_link = [ product.find_elements_by_class_name("core")[0].get_attribute('href') if
                                         len(product.find_elements_by_class_name("core")) > 0 else np.nan ]
                    for name in Name :
                        for price_Aft_dis in Price_After_disc :
                            for product_Bef_dis in Price_Before_disc :
                                for product_dis in Product_disc:
                                    for Product_eval in Product_evaluate :
                                        for Product_lin in Product_link :
                                            sleep(page_timeout)
                        
                                            writer.writerow({'Name' : name ,
                                                             'Price_After_disc' : price_Aft_dis ,
                                                             'Price_Before_disc' : product_Bef_dis , 
                                                             'Product_disc' : product_dis ,
                                                             'Product_evaluate' : Product_eval ,
                                                             'Product_link' : Product_lin ,
                                                             'Category' : cat})
                            
    
            except : 
                continue
    
    
    
```
