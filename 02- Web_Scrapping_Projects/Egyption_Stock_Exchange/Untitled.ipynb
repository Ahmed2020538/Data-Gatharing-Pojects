```python
import pandas as pd 
import requests as R 
import csv , json
from bs4 import BeautifulSoup as B

print("==============================( Loading Requiers Liberaries )=====================================")


password = int(input("Enter your Password :: "))
if password == 637 :
    print(f"Done")
else :
    print("Pass wrong")
    
print("==============================( Hello My Favor Client )=====================================")
    
def get_books_Name(url) :
    Books_name = []
    with open("Books_All_Details.csv" , "w+") as f :
        writer = csv.DictWriter( f , fieldnames=["Title" , "Rateing" , "Price" , "status" ])
        writer.writeheader()
        res = R.get(url)
        soup = B(res.text , "html.parser")
        data = soup.find_all("article"  , attrs={"class" : "product_pod"})
        for d in data :
            Title = d.find("h3").get_text()
            Books_name.append(Title)
            #print(f"Title :: {Title}\n")
    return Books_name

print("==============================( Books Name Loaded )=====================================")

def get_categories(url) :
    with open("Categories.csv" , "w+" ) as f :
        categories = {}
        writer = csv.DictWriter( f , fieldnames=["Category" , "Category_Link"])
        writer.writeheader()
        res = R.get(url)
        soup = B(res.text , "html.parser")
        data = soup.find("div" , attrs={"class": "side_categories"}).find("ul").find("li").find("ul").find_all("a")
        for d in data :
            Category = d.get_text().strip()
            Link = d.get("href").replace("../" , "http://books.toscrape.com/catalogue/category/books/")
            if Link != "index.html":
                categories[Category] = Link
                
                writer.writerow({"Category" : Category , "Category_Link" :Link})
                #print(f"Category :: {Category}\nLink :: {Link}")
        return categories
    
print("==============================( Categories and thier Url Loaded )=================================")
    
    
caties = get_categories("http://books.toscrape.com/catalogue/category/books/mystery_3/index.html")
with open ("Books_Data.csv" , "w+") as f :
    writer = csv.DictWriter(f , fieldnames=["Category_Name" , "Category_Link" , "Books_Name"])
    writer.writeheader()
    for cat_Name , Cat_Link in caties.items() :
        writer.writerow({"Category_Name" : cat_Name , "Category_Link" : Cat_Link, "Books_Name" : get_books_Name(Cat_Link)})
        print(f"cat_Name :: {cat_Name}\nCat_Link :: {Cat_Link}\n       =========================================================\nBooks\n--------\n{get_books_Name(Cat_Link)}")
    
print("==============================( All Books Data Extracted )=================================")
print("============================( Thanks for Use our Scraping verion )================================")
print(f"CopyWriter :: Ahmed Elsayed Ebrahim\n\nMy Number :: +201013900542\n\nMyLinkedIn Acount ::")

```
