```python
# Load Requires Liberaries ::
#-------------------------------
from selenium import webdriver
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager as CDM
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from openpyxl import load_workbook
from time import sleep
import csv
import sqlite3 as S
import pandas as pd
import numpy as np
import time as t
```

```python
url = "https://www.youtube.com/c/MustafaOthman/videos"
driver = webdriver.Chrome(CDM().install())
driver.get(url)
t.sleep(2)
print("WebSite loading done")
```

```python
with open("Mustafa_Abdelsalam_Data.csv" , "w+" , encoding = "utf-8") as f :
    writer = csv.DictWriter( f , fieldnames=["Title" , "Viewers" , "Publish_Time"])
    writer.writeheader()
    Data = driver.find_elements( By.ID, 'details')
    for d in Data :
        Vedio_Title = d.find_element(By.TAG_NAME , "h3").text
        Viewers_Count = d.find_element(By.ID , 'metadata-line').text.split("s")[0].strip().replace("view" ,
                                                                                               "views")
        Publish_Time = d.find_element(By.ID , 'metadata-line').text.split("s")[1].strip()
        writer.writerow({"Title" : Vedio_Title, "Viewers" : Viewers_Count, "Publish_Time" :Publish_Time})
        print(f"""Vedio_Title   :: {Vedio_Title}\nViewers_Count :: {Viewers_Count}
Publish_Time  ::{Publish_Time}\n=============""")
```
