```python
# Load Requires Liberaries ::
#-------------------------------
from selenium import webdriver
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager as CDM
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from openpyxl import load_workbook
from time import sleep
import csv , json , xlsxwriter
import sqlite3 as S
import pandas as pd
import numpy as np
```

```python
# For All Pages::
#----------------
All_data = []
try :
    for i in range(1 , 21 , 1 ) :
        #url = "https://webscraper.io/test-sites/e-commerce/static/computers/laptops?page=" + str(i)
        url = f"https://webscraper.io/test-sites/e-commerce/static/computers/laptops?page={i}" 
        driver = webdriver.Chrome(CDM().install())
        driver.get(url)
        print(f"\n================================( Page ({i})  Data )=============================")
        Product_Type     = driver.find_elements(By.CLASS_NAME , "title")
        Product_Describe = driver.find_elements(By.CLASS_NAME , "description")
        Product_Price    = driver.find_elements(By.CLASS_NAME , "caption")
        Product_Evalute  = driver.find_elements(By.CLASS_NAME , "ratings")
        
        for i in range(len(Product_Type)) :
            All_data.append([ Product_Type[i].text, Product_Describe[i].text, Product_Price[i].text.split(" ")[0].split(".")[0].split("$")[1],Product_Evalute[i].text.split(" ")[0]])
            print(f"""Product_Type :: {Product_Type[i].text}\nProduct_Describe :: {Product_Describe[i].text}
Product_Price :: {Product_Price[i].text.split(" ")[0].split(".")[0].split("$")[1]}\nProduct_Evalute :: {Product_Evalute[i].text.split(" ")[0]}\n==============""")
        print(f"\n=============================( Page Data Finshied )=====================================")
        # create xlsx file to insert data 
        with xlsxwriter.Workbook('All_Data.xlsx') as workbook:
            worksheet = workbook.add_worksheet()
            # set xlsx headers (Colins name)
            header_data  = ["Product_Type" ,  "Product_Describe" , "Product_Price" , "Product_Evalute"]
            for col_num, header in enumerate(header_data):
                worksheet.write(0, col_num, header)
            # wrie Data in xlsx file row by row
            for row_num, data in enumerate(All_data):
                worksheet.write_row(row_num +1, 0, data)
        print(f"\n================================(Insert Data In Xlsx Done Perfect)=================================")
                 
except :
    print(f"Page Error (404)")
    pass
driver.close()
```
