```python
import pandas as pd
import requests as R
import csv , json
from bs4 import BeautifulSoup as B
```

```python
def get_categories_links(url) :
    #url ="https://chroniclingamerica.loc.gov/titles/;start=A"
    Included_Links = []
    with open("Chronicling_America(Included).csv" , "w+") as f :
        writer = csv.DictWriter( f , fieldnames=["Includes_Links"])
        writer.writeheader()
        res = R.get(url)
        soup = B(res.text , "html.parser")
        Links = soup.find("table" , attrs = {"class" : "data"}).find_all("a")
        for Link in Links :
            lin = Link.get("href")
            last_href = "https://chroniclingamerica.loc.gov/" + lin
            Included_Links.append(last_href)
            #print(f"The Link ::\n{Included_Links}")
            writer.writerow({"Includes_Links" : Included_Links})
    return  Included_Links
```

```python
get_categories_links("https://chroniclingamerica.loc.gov/titles/;start=A")
```

```python
def category(url):
#url ="https://chroniclingamerica.loc.gov/titles/;start=A"
    with open("Chronicling_America.csv" , "w+") as f :
        writer = csv.DictWriter( f , fieldnames=["Litters" , "Litter_links"])
        writer.writeheader()
        categories ={}
        res = R.get(url)
        soup = B(res.text , "html.parser")
        Alphas = soup.find("div" , attrs = {"class" : "pagination pagination-mini"}).find("ul" , attrs = {"class" : "alpha_tags"}).find_all("a")
        for Alpha in Alphas :
            Litters = Alpha.get_text()
            lin = Alpha.get("href")
            last_href = "https://chroniclingamerica.loc.gov/" + lin
            categories[Litters] = last_href
            writer.writerow({"Litters" : Litters, "Litter_links" : last_href})
            #print(f"The Litter :: {Litters}\nThe Link ::\n{last_href}")
        return categories
```

```python
Cates = category("https://chroniclingamerica.loc.gov/titles/;start=A")
Cates
```
