```python
import pandas as pd
import csv, json , xlsxwriter
import seaborn as sn
import requests as R
from bs4 import BeautifulSoup as B
```

```python
def get_Amazo_data(url , j) :
    with open( "Amazon_data.csv" , "w+") as f :
        writer = csv.DictWriter( f , fieldnames=["" , "" , "" , ""])
        try :
            for i in range (j) :
                url = f"https://www.amazon.in/gp/bestsellers/books/ref=zg_bs_pg_1?ie=UTF8&pg={i}"
                res = R.get(url)
                soup = B(res.text , "html.parser")
                data = soup.find_all("div"  , attrs = {"class" : "p13n-sc-uncoverable-faceout"})
                for d in data :
                    Titel = d.find("div" , attrs = {"class" : "_cDEzb_p13n-sc-css-line-clamp-1_1Fn1y"}).get_text()
                    Art_Name = d.find("div" , attrs = {"class" : "a-row a-size-small"}).get_text()
                    Rate = d.find("span" , attrs = {"class" : "a-size-small"}).get_text()
                    Status = d.find("span" , attrs = {"class" : "a-size-small a-color-secondary a-text-normal"}).get_text()
                    Price = d.find("div" , attrs = {"class" : "_cDEzb_p13n-sc-price-animation-wrapper_3PzN2"})
                    print(f"Titel :: {Titel}\nArt_Name :: {Art_Name}\nRate :: {Rate}\nStatus :: {Status}\n======")
                    #print(f"Price :: {Price}")
                    
                print(f"Page Switched :: {i}\n----------------")
        except :
            print("No Page")
```

```python
get_Amazo_data("https://www.amazon.in/gp/bestsellers/books/ref=zg_bs_pg_1?ie=UTF8&pg=" , 2)
```
