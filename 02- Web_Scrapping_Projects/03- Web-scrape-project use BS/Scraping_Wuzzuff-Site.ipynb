```python
def get_WuZZuff_data() :
    with open("Data_Scientist.csv" , "w+") as f :
        writer = csv.DictWriter(f , fieldnames=["Jop_Title" , "Againt" , "Location" ,"Work_Case"  , "Experince_years" , "Jop_Describtion"])
        writer.writeheader()
        con =  S.connect("WuZZuff.db")
        con.execute("create table Data_Scientist ( Jop_Title str, Againt str, Location str, Work_Case str, Experince_years str, Jop_Describtion str)")
        try :
            for i in range(5) :
                url  = f"https://wuzzuf.net/search/jobs/?a=navbl&q=Data%20scientist&start={i}"
                res  = R.get(url)
                soup = B(res.text , "html.parser")
                data = soup.find_all("div" , attrs= {"class" : "css-1gatmva e1v1l3u10"})
                for d in data :
                    Jop_Title = d.find("h2").get_text()
                    Againt = d.find("a" , attrs={"class" : "css-17s97q8"}).get_text()
                    Location = d.find("span" , attrs={"class" : "css-5wys0k"}).get_text()
                    Work_Case = d.find("span" , attrs= {"class" , "css-1ve4b75 eoyjyou0"}).get_text()
                    Experince_years = d.find("div").get_text().split(" · ")[1]
                    Jop_Describtion = d.find("div").get_text().split("Exp ·")[1]
                    
                    writer.writerow({"Jop_Title" :Jop_Title , "Againt" : Againt , "Location" : Location , "Work_Case" : Work_Case , "Experince_years" : Experince_years , "Jop_Describtion" : Jop_Describtion} )
                    sql  = "insert into Data_Scientist (Jop_Title , Againt , Location , Work_Case , Experince_years , Jop_Describtion) values(?,?,?,?,?,?)"
                    data = [ Jop_Title , Againt,Location , Work_Case , Experince_years , Jop_Describtion]
                    con.execute(sql , data)
                    con.commit()
                    print(f"The Jop_Title :: {Jop_Title}\nThe Againt : {Againt}\nThe Location :: {Location}\nThe Work_Case :: {Work_Case}\nThe Experince_years :: {Experince_years}\nTheJop_Describtion :: {Jop_Describtion}\n{'*'*50}")
                print(f"{'='*50}\n{'-' *20} Page_Switched {'-'*20}\n{'='*50}")
                
        except :
            print("Page not Found")
            
            
            
        
```
